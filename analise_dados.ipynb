1 - Realizando os imports necessários.

```python
import pandas as pd
import matplotlib.pyplot as plt
from fpdf import FPDF

tabela = pd.read_csv("Dados_Graduacao_Brasil.csv")
tabela.info()
display(tabela)
```

2 - Criar um DataFrame e filtrar para região Sudeste.

```python
df = pd.DataFrame(tabela)

regiao = df[df['REGIAO'] == 'SUDESTE']

display(regiao)
```

3 - Filtrar os cursos apenas da região Sudeste. E somente o nome do curso, grau, cidade e modalidade.

```python
df1 = pd.DataFrame(tabela)

regiao = df1[df1['REGIAO'] == 'SUDESTE']

cursos = regiao[["NOME_CURSO", "GRAU", "MODALIDADE","UF", "REGIAO"]].groupby(["NOME_CURSO", "GRAU", "MODALIDADE", "UF", "REGIAO"]).sum()

display(cursos)
```

4 - Gráfico que demonstra a oferta de cursos por cidade da região Sudeste.

```python
qnt_cursos_por_uf = regiao['UF'].value_counts()

plt.title("Distribuição de cursos na região sudeste do Brasil")
plt.pie(qnt_cursos_por_uf, labels=qnt_cursos_por_uf.index, autopct="%1.2f%%")
plt.savefig("grafico_regiao.png")
plt.show()
```

5 - Filtar apenas os cursos da cidade de São Paulo.

```python
df2 = pd.DataFrame(tabela)

estado_sp = df2[df2['UF'] == 'SP']

cursos = estado_sp[["NOME_CURSO", "GRAU", "MODALIDADE", "UF"]].groupby(["NOME_CURSO", "GRAU", "MODALIDADE", "UF"]).sum()

display(cursos)
```

grau sp

```python
df3 = pd.DataFrame(tabela)

grau_cursos_sp = df3[df3['UF'] == 'SP']

cursos = grau_cursos_sp[["NOME_CURSO", "GRAU", "UF"]].groupby(["NOME_CURSO", "GRAU", "UF"]).sum()

display(cursos)
```

grafico grau

```python
grau_cursos_sp = estado_sp['GRAU'].value_counts()

plt.figure(figsize=(10, 6))
grau_cursos_sp.plot(kind='bar')
plt.title("Grau de especialização dos cursos no estado de São Paulo")
plt.xlabel("Grau de Especialização")
plt.ylabel("Quantidade de Cursos")
plt.xticks(rotation=0)  
plt.savefig("grafico_grau_sp.png")

plt.tight_layout()
plt.show()
```
